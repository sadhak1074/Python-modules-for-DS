```python
import numpy as np #printing common elements
n12=np.array([10, 20, 30, 40, 50, 60])
n13=np.array([50, 60, 70, 80, 90])
np.intersect1d(n12, n13)
```

```python
n12=np.array([10, 20, 30, 40, 50, 60]) #printing out ucommon elents of first array
n13=np.array([50, 60, 70, 80, 90])
np.setdiff1d(n12, n13)
```
